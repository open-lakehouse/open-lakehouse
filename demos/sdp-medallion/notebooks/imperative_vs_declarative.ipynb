## Same result, a third of the code

| | Imperative | Declarative (SQL) |
|---|---|---|
| Execution order | you write it | inferred from the graph |
| Reading upstream tables | explicit per step | a name in the `FROM` |
| Writes | one call per table | none - the `CREATE` *is* the table |
| Validation | at runtime | up front, before any data moves |
| Lines of code | ~90 | ~30 |

Declarative pipelines are shorter, validated before they run, and easier to hand to someone else - on fully open-source Spark.
