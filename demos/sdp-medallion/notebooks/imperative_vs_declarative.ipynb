## Side by side

| | Imperative (`imperative_medallion.py`) | Declarative (`transformations/*.sql`) |
|---|---|---|
| Execution order | you write it | inferred from the DAG |
| Reading upstreams | explicit `read.table()` per step | a name in the `FROM` |
| Writes | a `writeTo(...)` call per table | none — the `CREATE` *is* the table |
| Validation | runtime only | `dry-run` checks the whole graph up front |
| Lines of code | ~90 | ~30 |

> 💬 **Say to close:** "Same result, a third of the code, and the framework catches your mistakes before they cost you a job run. That's the case for declarative."
